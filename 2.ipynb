```python
#2.3.5
a, b = int(input()), int(input()) # ввести -1 в любое поле если надо прекратить
B = []
while (a != 100) or (b != 100):
    A = [x for x in list(x.groupby(['ЕГЭ...матем.']).count().index) if a < x and x <= b]
    B.append([str(a + 1)+' - '+str(b), x.groupby(['ЕГЭ...матем.']).count().loc[A].sum().iloc[1]])
    a = int(input())
    b = int(input())
y = pd.DataFrame(B,columns=['диапазон', 'число студентов'])
y
```

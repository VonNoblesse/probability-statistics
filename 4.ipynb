```python
import scipy
import math
import statistics as stat
import numpy as np
import matplotlib.pyplot as plt
```

```python
#4.1(a)(I)
n = 20
A = np.array([scipy.stats.bernoulli.rvs(0.5, size=1000) for i in range(n)]) # A_i_j = A[j][i]
S = np.sum(A,axis=0)
X = np.sum(S) / (1000 * n)
B = []
B = [B + a for a in A.tolist()][0]
S2 = stat.pstdev(B) ** 2
x = np.arange(0, n, 0.01)
plt.hist(S, histtype='step', cumulative=True, bins=len(S),weights=[0.001 for i in range(1000)])
plt.plot(x, scipy.stats.norm.cdf(x,n*X,np.sqrt(n * S2)))
plt.show()
```

```python
#4.3(1 вариант)(1)
n = 100
k = 1000
l = 5 #параметр
A = np.array([scipy.stats.poisson.rvs(l,size=k) for i in range(n)]) # A_i_j = A[j][i]
S = np.sum(A,axis=0) - l * n
a = np.arange(-100,100,1)
f = scipy.stats.norm.pdf(a,0,np.sqrt(n * l))
b = np.arange(min(S),max(S),1)
g = scipy.stats.poisson.pmf(b, mu=n*l, loc=-n*l)
figure, axis = plt.subplots(figsize=(5, 5))
axis.hist(S,density=True,alpha=0.5)
axis.plot(a,f,color='purple')
axis.plot(b,g,color='green')
axis.set_ylim(0,0.04)
axis.set_xlim([-100,100])
plt.show()
```

```python
#4.4 (bern)
n_list = [5,10,20,50,100,500]
figure, axis = plt.subplots(2, 3, figsize=(15, 10))
for i in range(len(n_list)):
    n = n_list[i]
    p = 1/3
    x = np.arange(-5,5,0.5)
    S = scipy.stats.binom.cdf((x+n*p/np.sqrt(n*p*(1-p)))*(np.sqrt(n*p*(1-p))),n,p)
    N = scipy.stats.norm.cdf(x)
    

    C = 0.4784
    E = p * (1-p)**3 + (1-p)*p**3 
    D = p * (1 - p)
    Berry1 = [C * E / (D**(3/2) * np.sqrt(n)) for i in x]
    Berry2 = [-C * E / (D**(3/2) * np.sqrt(n)) for i in x]
    Zero = [0 for i in x]

    a3 = ( p * (1-p)**3 - (1-p)*p**3 ) / D ** (3 / 2)
    decomp = np.array([a3 / (6 * math.sqrt(2 * math.pi * n)) * (1 - t * t) * math.exp(- t * t / 2) for t in x])
    
    if i > 2:
        j = 1
    else:
        j = 0
    axis[j,i%3].bar(x,S-N,color='gray',width=0.4)
    axis[j,i%3].plot(x, Berry1,color='red')
    axis[j,i%3].plot(x, Berry2,color='red')
    axis[j,i%3].plot(x, Zero,color='black')
    axis[j,i%3].plot(x, decomp)
plt.show()
```

```python
#4.4 (bern)
n_list = [5,10,20,50,100,500]
figure, axis = plt.subplots(2, 3, figsize=(15, 10))
for i in range(len(n_list)):
    n = n_list[i]
    l = 5
    x = np.arange(-5,5,0.5)
    C = 0.4784
    E = n*(1/l)
    D = n*(1/l)**2
    S = scipy.stats.gamma.cdf(x*np.sqrt(n*D)+E,n,scale=1/l)
    N = scipy.stats.norm.cdf(x)
    
    E3 = (18 - 2 * math.exp(1))/(math.exp(1) * l**4)
    Berry1 = [C * E3 / (D**(3/2) * np.sqrt(n)) for i in x]
    Berry2 = [-C * E3 / (D**(3/2) * np.sqrt(n)) for i in x]
    Zero = [0 for i in x]
    a3 = E3/(1/l)**(3/2)
    decomp = np.array([a3 / (6 * math.sqrt(2 * math.pi * n)) * (1 - t * t) * math.exp(- t * t / 2) for t in x])
    
    if i > 2:
        j = 1
    else:
        j = 0
    axis[j,i%3].bar(x,S-N,color='gray',width=0.4)
    axis[j,i%3].plot(x, Berry1,color='red')
    axis[j,i%3].plot(x, Berry2,color='red')
    axis[j,i%3].plot(x, Zero,color='black')
    axis[j,i%3].plot(x, decomp)
plt.show()
```

```python
import seaborn as sb
import scipy
import math
import statistics as stat
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#9.1(a)
teta0 = 1/2
teta1 = 1/3
n = 10
C = np.arange(0, n, 0.5)

teta_true = teta0
x1 = 1 - scipy.stats.binom.cdf(C, n, teta_true) #  I 

teta_true = teta1
x2 = scipy.stats.binom.cdf(C, n, teta_true) #  II 

teta_true = teta1
x3 = 1 - scipy.stats.binom.cdf(C, n, teta_true) # 

fig, ax = plt.subplots(1,3, figsize=(15,5))
ax[0].plot(C,x1)
ax[0].bar(C,x1,alpha=0.2,color='green',width=0.5)
ax[0].set_title(' I ')
ax[1].plot(C,x2)
ax[1].bar(C,x2,alpha=0.2,color='green',width=0.5)
ax[1].set_title(' II ')
ax[2].plot(C,x3)
ax[2].bar(C,x3,alpha=0.2,color='green',width=0.5)
ax[2].set_title('')
plt.show()
fig, ax = plt.subplots(1,1,figsize = (10,10))
ax.plot(C,x1,label=' I ')
ax.plot(C,x2,label=' II ')
ax.plot(C,x3,label='')
ax.legend()
plt.show()
```

```python
#9.2(a)(a)
teta0 = 0
teta1 = 1
m = 100
n = 10
p_value_a_list = []
p_value_b_list = []
p_value_c_list = []
teta = teta0
for i in range(m):
    x = scipy.stats.norm.rvs(teta,1,size=n)
    x = np.sort(x)
    y = x.sum()
    p_value_a = 1 - scipy.stats.norm.cdf(y,0,np.sqrt(n))
    p_value_b = scipy.stats.norm.cdf(y,0,np.sqrt(n))
    
    z = abs(np.median(x))
    x0 = scipy.stats.norm.ppf(0.5,loc=0,scale=1)
    sigma = 1/(2 * scipy.stats.norm.pdf(x0,loc=0,scale=1))
    p_value_c = 2 - 2 *scipy.stats.norm.cdf(np.sqrt(n)*z,loc=0,scale=sigma)
    
    p_value_a_list.append(p_value_a)
    p_value_b_list.append(p_value_b)
    p_value_c_list.append(p_value_c)
p_value_a_list = np.sort(p_value_a_list)
p_value_b_list = np.sort(p_value_b_list)
p_value_c_list = np.sort(p_value_c_list)
y = [float(i+1)/m for i in range(m)]
fig, ax = plt.subplots(1,1,figsize=(10,10))
ax.plot(p_value_a_list,y,label='teta1 > 0')
ax.plot(p_value_b_list,y,label='teta1 < 0')
ax.plot(p_value_c_list,y,label='median')
ax.plot(y,y,color='black')
ax.set_title('teta = '+str(teta))
ax.legend()
plt.show()
```

```python
#9.3
n_list = [10,20,50,100]
teta0 = 1/2
teta1 = 1/3
alpha = 0.05
fig, ax = plt.subplots(1,len(n_list),figsize=(20,5))
for i in range(len(n_list)):
    n = n_list[i]
    C = np.arange(0,n,0.5)
    
    teta_true = teta0
    x1 = scipy.stats.binom.cdf(C, n, teta_true) #  I 

    teta_true = teta1
    x2 = 1 - scipy.stats.binom.cdf(C, n, teta_true) #  II 
    
    m = min([j for j in range(len(C)) if (x2[j] <= alpha)])
    m = round(x2[m],3)
    ax[i].set_title('n = '+str(n)+'; intersection = '+str(m))
    ax[i].plot(C,x1,label='I')
    ax[i].plot(C,x2,label='II')
    ax[i].legend()
print('alpha =',alpha)
plt.show()
```

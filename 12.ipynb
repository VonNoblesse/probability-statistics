```python
#12.1(N[0,1])(2)
n_list = [50]
a = np.arange(-2,2,0.1)
fig, ax = plt.subplots(1,2,figsize=(14,7))
theor = scipy.stats.norm.cdf(a)
ax[0].plot(a,theor,color='black')
ax[1].plot(a,theor,color='black')
for i in range(len(n_list)):
    n = n_list[i]
    x = []
    x = scipy.stats.norm.rvs(size=n)
    F = ECDF(x)
    y = F(a)
    K = scipy.stats.kstwobign.ppf(0.95)
    ax[0].plot(a,y)
    ax[1].plot(a,y)
    ax[0].fill_between(a, theor - K/np.sqrt(n), theor + K/np.sqrt(n), color = 'green', alpha = 0.2, label = "kolmogorov")
    y1 = np.sqrt(math.log(2./0.05)/(2*n))
    ax[1].fill_between(a, theor - y1, theor + y1, color = 'orange', alpha = 0.2, label = "ineq")
ax[0].legend()
ax[1].legend()
plt.show()
```

```python
#12.2(a)
n_list = [50,100,500]
mu = -3
sigma2 = 8
alpha = 0.05
fig, ax = plt.subplots(1,1,figsize=(7,7))
for i in range(len(n_list)):
    n = n_list[i]
    x = (scipy.stats.norm.rvs(mu,np.sqrt(sigma2),size=n)-mu)/np.sqrt(sigma2)
    p_value = scipy.stats.kstest(x,'norm').pvalue
    if p_value <= alpha:
        print(' n =',n,'   ')
    else:
        print(' n =',n,'   ')
    P = [p_value]
    k = 99
    for j in range(k):
        x = (scipy.stats.norm.rvs(mu,np.sqrt(sigma2),size=n)-mu)/np.sqrt(sigma2)
        p_value = scipy.stats.kstest(x,'norm').pvalue
        P.append(p_value)
    P = np.sort(P)
    k_list = [float(l+1)/(k+1) for l in range(k+1)]
    ax.plot(P,k_list)
    ax.plot(k_list,k_list,color='black',alpha=0.3)
```

```python
#12.3
m = 100
P = []
n = 1000
mu = 2
sigma2 = 4
for i in range(m):
    x = scipy.stats.norm.rvs(mu,np.sqrt(sigma2),size=n)
    mean = x.mean()
    var = np.sqrt(stat.variance(x))
    x = (x - mean)/var
    p_value = scipy.stats.kstest(x,'norm').pvalue
    P.append(p_value)
P = np.sort(P)
m_list = [float(i+1)/m for i in range(m)]
fig, ax = plt.subplots(1,1,figsize=(7,7))
ax.plot(m_list,P)
ax.plot(m_list,m_list,color='black')
plt.show()
```

```python
#3.2(1)
X = scipy.stats.binom.rvs(100, 0.04, size=1000)
Y = scipy.stats.poisson.rvs(4, size=1000)
plt.bar(np.array(X), [len([1 for y in X if y == x]) for x in X])
plt.show()
plt.bar(np.array(Y), [len([1 for y in Y if y == x]) for x in Y])
plt.show()
width = 0.4
plt.bar(np.array(X)-width/2, [len([1 for y in X if y == x]) for x in X], width=width)
plt.bar(np.array(Y)+width/2, [len([1 for y in Y if y == x]) for x in Y], width=width)
plt.show()
```

```python
#3.3
x = pd.read_csv(r"Priem.csv")
y1 = x[['ЕГЭ...матем.', 'ЕГЭ...рус.яз.']].groupby(['ЕГЭ...матем.']).count().reset_index()
plt.bar(y1['ЕГЭ...матем.'],y1['ЕГЭ...рус.яз.'])
plt.title('матем')
plt.show()
y2 = x[['ЕГЭ...матем.', 'ЕГЭ...рус.яз.']].groupby(['ЕГЭ...рус.яз.']).count().reset_index()
plt.bar(y2['ЕГЭ...рус.яз.'],y2['ЕГЭ...матем.'])
plt.title('рус.яз')
plt.show()
width = 0.4
plt.figure(figsize=(20,5))
plt.bar(np.array(y1['ЕГЭ...матем.'])-width/2,y1['ЕГЭ...рус.яз.'],width=width)
plt.bar(np.array(y2['ЕГЭ...рус.яз.'])+width/2,y2['ЕГЭ...матем.'],width=width)
plt.show()
```

```python
#3.8
import random
from pandas.plotting import parallel_coordinates
x = sb.load_dataset('titanic')
indexes = [random.randint(0, x.shape[0]) for i in range(30)]
x1 = pd.DataFrame([x.loc[i] for i in indexes],columns=x.columns)
maxage = max(x1['age'])
x1['who'] = x1['who'].replace(['man','woman','child'],[0, maxage / 2, maxage])
x1['pclass'] = x1['pclass'].replace([1,2,3],[0, maxage / 2, maxage])
x1['survived'] = x1['survived'].replace([0,1],[0,1])
x1['index'] = indexes
plt.figure(figsize=(10,10))
parallel_coordinates(x1, 'survived', cols=['pclass', 'age', 'who'],colormap='rainbow')
```

```python
import numpy as np
import scipy 
import matplotlib.pyplot as plt
```

```python
#1 (a)
#параметры
c = 2
n = 14
N = 100
p = 1/2
xi = np.zeros(N)


#вычисляем
for i in range(N):
    s = 0
    while s < c*2**n:
        x = scipy.stats.geom.rvs(p=p)+1
        s += x
        if x > n and s < c*2**n:
            xi[i] += 1

plt.hist(xi,density=True)
plt.show()
```

```python
#2 (a) n = 50
#параметры
n = 50
N = 1000
counts = np.zeros(N)

#вычисляем
for i in range(N):
    cells = np.zeros(n)
    for j in range(n):
        cell = scipy.stats.randint.rvs(0,50)
        cells[cell] += 1
    counts[i] = np.sum(cells == 0)
a = n/n
counts = (counts - n * np.exp(-a))/(np.sqrt(n*np.exp(-a)*(1-np.exp(-a))))
plt.hist(counts,density=True,bins=25)
plt.show()
```

```python
#4 (a) Теорема Петрова, лекция 8
#параметры
l = 2
n = 20
N = 1000

#моделируем
count = 0
for i in range(N):
    x = scipy.stats.expon.rvs(scale=1/l,size=n)
    S = np.cumsum(x)
    if S[-1] > 2 * n / 3:
        count += 1
pr = count / N
print(pr)
```

```python
#5 (a)(a)
n = 100
N = 1000
limit = np.zeros(N,dtype=object)


for j in range(N):
    E = 0
    D = 0
    S = 0
    for i in range(1,n+1):
        X = scipy.stats.multinomial.rvs(n=1,p=[1/(2*i*i),1-1/(i*i),1/(2*i*i)])
        index = np.where(X == 1)[0][0]
        if index == 0:
             S += -i
        elif index == 1:
            S += 0
        else:
            S += i
        D += 1
    limit[j] = (S-E)/np.sqrt(D)
plt.hist(limit,density=True,bins=50)
plt.show()
```
